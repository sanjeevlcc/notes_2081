```python
import numpy as np
import matplotlib.pyplot as plt

# let img1 be an image with no features
img1 = np.array([np.array([200, 200]), np.array([200, 200])])
img2 = np.array([np.array([200, 200]), np.array([0, 0])])
img3 = np.array([np.array([200, 0]), np.array([200, 0])])

kernel_horizontal = np.array([np.array([2, 2]), np.array([-2, -2])])
print(kernel_horizontal, 'is a kernel for detecting horizontal edges')

kernel_vertical = np.array([np.array([2, -2]), np.array([2, -2])])
print(kernel_vertical, 'is a kernel for detecting vertical edges')

# We will apply the kernels on the images by
# elementwise multiplication followed by summation
def apply_kernel(img, kernel):
    return np.sum(np.multiply(img, kernel))

# Visualizing img1
plt.imshow(img1)
plt.axis('off')
plt.title('img1')
plt.show()

# Checking for horizontal and vertical features in image1
print('Horizontal edge confidence score:', apply_kernel(img1,
                                            kernel_horizontal))
print('Vertical edge confidence score:', apply_kernel(img1,
                                            kernel_vertical))

# Visualizing img2
plt.imshow(img2)
plt.axis('off')
plt.title('img2')
plt.show()

# Checking for horizontal and vertical features in image2
print('Horizontal edge confidence score:', apply_kernel(img2,
                                            kernel_horizontal))
print('Vertical edge confidence score:', apply_kernel(img2,
                                            kernel_vertical))

# Visualizing img3
plt.imshow(img3)
plt.axis('off')
plt.title('img3')
plt.show()

# Checking for horizontal and vertical features in image3
print('Horizontal edge confidence score:', apply_kernel(img3,
                                            kernel_horizontal))
print('Vertical edge confidence score:', apply_kernel(img3,
                                            kernel_vertical))
```

```python

# Basic Matrix Convolution and Edge Detection Lab

import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Step 1: Define a simple 2D grayscale image (5x5 matrix)
image = np.array([
    [10, 10, 10, 10, 10],
    [10, 50, 50, 50, 10],
    [10, 50, 100, 50, 10],
    [10, 50, 50, 50, 10],
    [10, 10, 10, 10, 10]
], dtype=np.float32)

print("Original Image:\n", image)

# Step 2: Define a simple kernel (Edge detection - Laplacian)
laplacian_kernel = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])

# Step 3: Perform convolution using scipy.ndimage.convolve
convolved_image = ndimage.convolve(image, laplacian_kernel)

print("After Convolution (Laplacian filter):\n", convolved_image)

# Step 4: Visualize the images
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.colorbar()

plt.subplot(1,2,2)
plt.title("Edge Detected Image (Laplacian)")
plt.imshow(convolved_image, cmap='gray')
plt.colorbar()

plt.show()

# Step 5: Sobel Edge Detection (Horizontal and Vertical)

sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

sobel_y = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])

edge_x = ndimage.convolve(image, sobel_x)
edge_y = ndimage.convolve(image, sobel_y)

# Calculate magnitude of gradient
edge_magnitude = np.hypot(edge_x, edge_y)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Sobel Edge X")
plt.imshow(edge_x, cmap='gray')
plt.colorbar()

plt.subplot(1,3,2)
plt.title("Sobel Edge Y")
plt.imshow(edge_y, cmap='gray')
plt.colorbar()

plt.subplot(1,3,3)
plt.title("Edge Magnitude")
plt.imshow(edge_magnitude, cmap='gray')
plt.colorbar()

plt.show()
```
