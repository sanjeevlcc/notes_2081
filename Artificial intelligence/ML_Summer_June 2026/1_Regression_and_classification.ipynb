# REGRESSION VS CLASSIFICATION
## Notes with Data and Examples

---

### 1. INTRODUCTION

* Regression and classification are two major supervised machine learning techniques.
* Both use input data to predict an output.
* The main difference is the type of output.
* **Regression** predicts a numerical value.
* **Classification** predicts a category or class.

---

### 2. SIMPLE DIFFERENCE

* **Regression:** Predicts *"How much?"* or *"What value?"*
* **Classification:** Predicts *"Which class?"* or *"Which category?"*

---

### 3. EXAMPLES

#### Example Set 1
* **Regression:** Predict student CGPA.
* **Classification:** Predict whether student performance is *Low*, *Medium*, or *High*.

#### Example Set 2
* **Regression:** Predict house price.
* **Classification:** Predict whether house is *Cheap*, *Medium*, or *Expensive*.

#### Example Set 3
* **Regression:** Predict sales revenue.
* **Classification:** Predict whether sales performance is *Poor* or *Good*.

---

### 4. REGRESSION

* Regression is used when the target variable is **continuous numerical data**.
* Continuous data can take many possible values.

#### Examples of Continuous Data
* 2.45
* 3.17
* 45000
* 12500.75
* 98.6

#### Regression Questions
* What will be the student CGPA?
* What will be the sales amount?
* What will be the crop yield?
* What will be the temperature?
* What will be the house price?

---

### 5. CLASSIFICATION

* Classification is used when the target variable is a **category or label**.
* The output belongs to a fixed, predefined group.

#### Examples of Categories
* Pass / Fail
* Spam / Not Spam
* Disease / No Disease
* Low / Medium / High
* Yes / No

#### Classification Questions
* Will the student pass or fail?
* Is the email spam or not?
* Is irrigation required or not?
* Is the customer likely to buy or not?
* Is the student performance Low, Medium, or High?

---

### 6. MAIN DIFFERENCE TABLE


| Basis | Regression | Classification |
| :--- | :--- | :--- |
| **Output Type** | Numerical value | Category or class |

| **Target Variable** | Continuous | Categorical |

### 7. REGRESSION EXAMPLE: STUDENT CGPA

* **Problem:** Predict CGPA using study hours and attendance.
* **Input Variables:** Study Hours, Attendance
* **Target Variable:** CGPA

#### Sample Data

| Student | Study Hours | Attendance | CGPA |
| :---: | :---: | :---: | :---: |
| S1 | 1 | 55 | 2.10 |
| S2 | 2 | 65 | 2.40 |
| S3 | 3 | 72 | 2.80 |
| S4 | 4 | 80 | 3.20 |
| S5 | 5 | 88 | 3.60 |
| S6 | 6 | 95 | 3.90 |

#### Regression Equation
$$\text{CGPA} = b_0 + b_1(\text{Study Hours}) + b_2(\text{Attendance})$$

**Example Setup:**
$$\text{CGPA} = 1.20 + 0.25(\text{Study Hours}) + 0.01(\text{Attendance})$$

#### Prediction Calculation
* **New Student Input:** Study Hours = 4, Attendance = 82
* **Calculation:**
  $$\text{CGPA} = 1.20 + 0.25(4) + 0.01(82)$$
  $$\text{CGPA} = 1.20 + 1.00 + 0.82$$
  $$\text{CGPA} = 3.02$$
* **Regression Output:** Predicted CGPA = **3.02**

> 📌 **Important Point:** The output is a number. Therefore, this is regression.

### 17. IMPORTANT NOTE ABOUT LOGISTIC REGRESSION

* ⚠️ **Logistic Regression has the word "regression" in its name, but it is used for classification.**

#### Example Targets
* Pass / Fail
* Spam / Not Spam
* Yes / No

#### How it Works
* **Output:** A raw probability score scaled tightly between `0` and `1`.
  * *Example:* Probability of Pass = **0.87**
* **Decision Boundary Rule:**
  * If $\text{Probability} > 0.50 \rightarrow$ **Predict Pass**
  * If $\text{Probability} \le 0.50 \rightarrow$ **Predict Fail**

---

### 18. REGRESSION EVALUATION METRICS

#### A. MAE (Mean Absolute Error)
* Measures the simple average magnitude of absolute prediction errors.
* *Example:* $\text{MAE} = 0.15$
* *Meaning:* The average model prediction misses the true student CGPA by $0.15$ units.
* 📌 **Lower MAE is better.**

#### B. RMSE (Root Mean Squared Error)
* Measures average error but squares variances before averaging to penalize outliers.
* *Meaning:* Gives an exponentially larger penalty to extreme errors.
* 📌 **Lower RMSE is better.**

#### C. R-SQUARED ($R^2$)
* Measures the total proportion of target variance explained by input model features.
* *Example:* $R^2 = 0.80$
* *Meaning:* $80\%$ of the structural variations in the target output are accurately explained by inputs.
* 📌 **Higher $R^2$ is better.**

---

### 19. CLASSIFICATION EVALUATION METRICS

#### A. ACCURACY
* Measures the total raw percentage of correctly determined classification instances.
* **Formula:**
  $$\text{Accuracy} = \frac{\text{Correct Predictions}}{\text{Total Predictions}}$$
* *Example:* $\text{Accuracy} = 90\%$ ($90$ out of $100$ cases were right).

#### B. PRECISION
* Measures structural accuracy among cases predicted as positive.
* *Example:* Model flags $20$ students to pass. Only $18$ actually do.
* **Calculation:**
  $$\text{Precision} = \frac{18}{20} = 90\%$$

#### C. RECALL
* Measures structural coverage among all actual positive cases available.
* *Example:* There are $25$ true passing students. Model catches $20$ of them.
* **Calculation:**
  $$\text{Recall} = \frac{20}{25} = 80\%$$

#### D. F1 SCORE
* Delivers the harmonic mean score to balance Precision and Recall together.
* 📌 **Higher F1 Score is better.**

#### E. CONFUSION MATRIX
* A cross-tabulation table indicating exact intersections of True vs. Predicted classes.


| | Predicted Pass | Predicted Fail |
| :--- | :---: | :---: |
| **Actual Pass** | 40 (True Positive) | 5 (False Negative) |
| **Actual Fail** | 3 (False Positive) | 32 (True Negative) |

### 22. WHEN TO USE REGRESSION

Use regression modeling when:
* The target output is strictly numerical.
* An exact, fine-grained value is required.
* The target variable can take infinitely many continuous values.

#### Standard Use Cases
* Sales revenue forecasting
* Academic CGPA evaluation
* Real estate house prices
* Agricultural crop yields
* Ambient temperature tracking
* Monthly household income

---

### 23. WHEN TO USE CLASSIFICATION

Use classification modeling when:
* The target output is a distinct label or category.
* The target features fixed, discrete classes.
* A definitive bucket-based decision is required.

#### Standard Use Cases
* Pass or Fail flags
* Spam or Not Spam filtering
* Buyer or Non-Buyer conversion tracking
* Irrigation Required or Not Required automation
* Disease or No Disease clinical screening
* Low, Medium, or High priority ranking

---

### 24. QUICK IDENTIFICATION METHOD

To find the right approach, analyze the output data properties:

* **Is the target output an explicit number?**
  * *Yes:* Use **Regression**
* **Is the target output a distinctive category?**
  * *Yes:* Use **Classification**

#### Core Context Comparison
* Target: Sales Revenue of *Rs. 18,500* $\rightarrow$ **Regression**
* Target: Sales Performance Tier *High* $\rightarrow$ **Classification**

### 26. SHORT EXAM ANSWER

* Regression and classification are two main categories of **supervised machine learning**.
* **Regression** predicts continuous numerical outputs (e.g., student CGPA, sales revenue, crop yield, or house prices).
* **Classification** predicts discrete categorical outputs (e.g., Pass/Fail, Spam/Not Spam, or Low/Medium/High groupings).
* Regression performance is validated using **MAE**, **RMSE**, and **$R^2$**.
* Classification performance is validated using **Accuracy**, **Precision**, **Recall**, **F1-Score**, and **Confusion Matrices**.
* *Example:* Estimating a numerical CGPA score is regression, while labeling it into Low, Medium, or High categories is classification.

# Regression and Classification Using Student Performance Data

## Project Objective

This notebook demonstrates the difference between:

1. **Regression**
2. **Classification**

The same student dataset will be used for both models.

## Regression Problem

The regression model predicts the exact numerical value of a student's CGPA.

```text
Input Variables → Study Hours, Attendance, Motivation, Previous GPA
Output Variable → CGPA

````python



# Cell 2 — Code Cell


# ============================================================
# STEP 1: IMPORT REQUIRED LIBRARIES
# ============================================================

# NumPy is used for numerical operations and random data generation.
import numpy as np

# Pandas is used for creating and managing tabular data.
import pandas as pd

# Matplotlib is used for graphs and data visualization.
import matplotlib.pyplot as plt

# Train-test split divides data into training and testing datasets.
from sklearn.model_selection import train_test_split

# Linear Regression predicts a continuous numerical value.
from sklearn.linear_model import LinearRegression

# Logistic Regression predicts classes or categories.
from sklearn.linear_model import LogisticRegression

# StandardScaler places input variables on a similar scale.
from sklearn.preprocessing import StandardScaler

# Pipeline combines preprocessing and model training.
from sklearn.pipeline import Pipeline

# Regression evaluation metrics.
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# Classification evaluation metrics.
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

# Set random seed so that we get the same results every time.
np.random.seed(42)

print("Libraries imported successfully.")
````

## Dataset Design

This notebook creates a synthetic dataset of 2,000 students.

### Input Variables

| Variable | Meaning | Type |
|---|---|---|
| study_hours | Average study hours per day | Numerical |
| attendance | Class attendance percentage | Numerical |
| motivation | Motivation score from 1 to 5 | Numerical |
| previous_gpa | Previous academic GPA | Numerical |
| internet_access | Internet availability | Binary |
| sleep_hours | Average sleep hours | Numerical |

### Regression Target

```text
CGPA

````python


# Cell 4 — Code Cell


# ============================================================
# STEP 2: CREATE A 2000-ROW STUDENT DATASET
# ============================================================

# Total number of student records.
n = 2000

# Create student IDs.
student_id = [
    f"STU_{i:04d}"
    for i in range(1, n + 1)
]

# Average study hours per day.
# Values are limited between 0.5 and 8 hours.
study_hours = np.clip(
    np.random.normal(
        loc=3.8,
        scale=1.4,
        size=n
    ),
    0.5,
    8
).round(2)

# Attendance percentage.
# Values are limited between 40% and 100%.
attendance = np.clip(
    np.random.normal(
        loc=78,
        scale=12,
        size=n
    ),
    40,
    100
).round(1)

# Motivation score between 1 and 5.
motivation = np.clip(
    np.random.normal(
        loc=3.3,
        scale=0.9,
        size=n
    ),
    1,
    5
).round(1)

# Previous GPA between 1.5 and 4.0.
previous_gpa = np.clip(
    np.random.normal(
        loc=2.8,
        scale=0.5,
        size=n
    ),
    1.5,
    4.0
).round(2)

# Internet access:
# 1 means Yes
# 0 means No
internet_access = np.random.choice(
    [0, 1],
    size=n,
    p=[0.25, 0.75]
)

# Average sleep hours.
sleep_hours = np.clip(
    np.random.normal(
        loc=6.8,
        scale=1.0,
        size=n
    ),
    4,
    9
).round(1)

print("Input variables generated successfully.")
````

## Creating the Regression Target

The CGPA is generated using a sample mathematical relationship.

```text
CGPA =
Constant
+ Effect of Study Hours
+ Effect of Attendance
+ Effect of Motivation
+ Effect of Previous GPA
+ Effect of Internet Access
+ Effect of Sleep Hours
+ Random Error

````python


# Cell 6 — Code Cell


# ============================================================
# STEP 3: CREATE THE NUMERICAL TARGET VARIABLE: CGPA
# ============================================================

# Random error represents unknown variables and natural variation.
random_error = np.random.normal(
    loc=0,
    scale=0.18,
    size=n
)

# Create CGPA using a linear relationship.
cgpa = (
    0.35
    + 0.18 * study_hours
    + 0.010 * attendance
    + 0.12 * motivation
    + 0.47 * previous_gpa
    + 0.10 * internet_access
    + 0.035 * sleep_hours
    + random_error
)

# Limit CGPA between 1.5 and 4.0.
cgpa = np.clip(
    cgpa,
    1.5,
    4.0
).round(2)

print("CGPA target created successfully.")
````

## Creating the Classification Target

CGPA will be converted into three performance classes.

```text
CGPA below 2.80           → Low
CGPA from 2.80 to 3.49    → Medium
CGPA 3.50 and above       → High

````python



# Cell 8 — Code Cell


# ============================================================
# STEP 4: CONVERT CGPA INTO PERFORMANCE CLASSES
# ============================================================

performance_class = pd.cut(
    cgpa,
    bins=[0, 2.79, 3.49, 4.0],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)

print("Performance classes created successfully.")
````

````python
# ============================================================
# STEP 5: CREATE THE FINAL DATAFRAME
# ============================================================

df = pd.DataFrame({
    "student_id": student_id,
    "study_hours": study_hours,
    "attendance": attendance,
    "motivation": motivation,
    "previous_gpa": previous_gpa,
    "internet_access": internet_access,
    "sleep_hours": sleep_hours,
    "cgpa": cgpa,
    "performance_class": performance_class
})

# Display the first 10 rows.
df.head(10)
````

# Part A: Regression Model

## Regression Objective

Predict the exact numerical CGPA using:

- Study hours
- Attendance
- Motivation
- Previous GPA
- Internet access
- Sleep hours

## Target

```CGPA

````python


# Cell 24 — Code Cell


# ============================================================
# STEP 15: PREPARE REGRESSION INPUT AND OUTPUT
# ============================================================

feature_columns = [
    "study_hours",
    "attendance",
    "motivation",
    "previous_gpa",
    "internet_access",
    "sleep_hours"
]

# X contains input variables.
X_regression = df[feature_columns]

# y contains the numerical target.
y_regression = df["cgpa"]

print("Regression Input Shape:", X_regression.shape)
print("Regression Output Shape:", y_regression.shape)
````

````python
# ============================================================
# STEP 16: SPLIT DATA FOR REGRESSION
# ============================================================

X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_regression,
    y_regression,
    test_size=0.20,
    random_state=42
)

print("Training Records:", X_reg_train.shape[0])
print("Testing Records:", X_reg_test.shape[0])
````

````python
# ============================================================
# STEP 17: CREATE AND TRAIN LINEAR REGRESSION MODEL
# ============================================================

regression_model = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

regression_model.fit(
    X_reg_train,
    y_reg_train
)

print("Regression model trained successfully.")
````

````python
# ============================================================
# STEP 18: MAKE REGRESSION PREDICTIONS
# ============================================================

regression_predictions = regression_model.predict(
    X_reg_test
)

print("First 10 Regression Predictions:")
print(regression_predictions[:10])
````

````python
# ============================================================
# STEP 19: EVALUATE REGRESSION MODEL
# ============================================================

regression_mae = mean_absolute_error(
    y_reg_test,
    regression_predictions
)

regression_rmse = mean_squared_error(
    y_reg_test,
    regression_predictions
) ** 0.5

regression_r2 = r2_score(
    y_reg_test,
    regression_predictions
)

print("REGRESSION MODEL RESULTS")
print("========================")
print("MAE :", round(regression_mae, 4))
print("RMSE:", round(regression_rmse, 4))
print("R²  :", round(regression_r2, 4))
````

RMSE

Lower RMSE is better.
R² Score

# Part B: Classification Model

## Classification Objective

Predict the student's performance class.

````python


# Cell 33 — Code Cell


# ============================================================
# STEP 22: PREPARE CLASSIFICATION INPUT AND OUTPUT
# ============================================================

X_classification = df[feature_columns]

y_classification = (
    df["performance_class"]
    .astype(str)
)

print("Classification Input Shape:", X_classification.shape)
print("Classification Output Shape:", y_classification.shape)
````

````python
# ============================================================
# STEP 23: SPLIT DATA FOR CLASSIFICATION
# ============================================================

X_cls_train, X_cls_test, y_cls_train, y_cls_test = train_test_split(
    X_classification,
    y_classification,
    test_size=0.20,
    random_state=42,
    stratify=y_classification
)

print("Classification Training Records:", X_cls_train.shape[0])
print("Classification Testing Records:", X_cls_test.shape[0])
````

````python
# ============================================================
# STEP 24: CREATE AND TRAIN LOGISTIC REGRESSION CLASSIFIER
# ============================================================

classification_model = Pipeline([
    ("scaler", StandardScaler()),
    (
        "model",
        LogisticRegression(
            max_iter=2000,
            random_state=42
        )
    )
])

classification_model.fit(
    X_cls_train,
    y_cls_train
)

print("Classification model trained successfully.")
````

````python
# ============================================================
# STEP 25: MAKE CLASSIFICATION PREDICTIONS
# ============================================================

classification_predictions = classification_model.predict(
    X_cls_test
)

print("First 10 Classification Predictions:")
print(classification_predictions[:10])
````

````python
# ============================================================
# STEP 26: EVALUATE CLASSIFICATION MODEL
# ============================================================

classification_accuracy = accuracy_score(
    y_cls_test,
    classification_predictions
)

classification_precision = precision_score(
    y_cls_test,
    classification_predictions,
    average="weighted",
    zero_division=0
)

classification_recall = recall_score(
    y_cls_test,
    classification_predictions,
    average="weighted",
    zero_division=0
)

classification_f1 = f1_score(
    y_cls_test,
    classification_predictions,
    average="weighted",
    zero_division=0
)

print("CLASSIFICATION MODEL RESULTS")
print("============================")
print("Accuracy :", round(classification_accuracy, 4))
print("Precision:", round(classification_precision, 4))
print("Recall   :", round(classification_recall, 4))
print("F1 Score :", round(classification_f1, 4))
````

## Classification Result Interpretation

### Accuracy

Accuracy shows the percentage of all correct predictions.

### Precision

Precision shows how many predicted class cases were correct.

### Recall

Recall shows how many actual class cases were identified.

### F1 Score

F1 Score balances precision and recall.

```text
Higher Accuracy, Precision, Recall and F1 Score are generally better.

# Input and Output Demonstration

The following examples give new student input values to both models.

The regression model predicts:

```text
Exact CGPA

The classification model predicts:

Low, Medium, or High Performance

````python


# Cell 44 — Code Cell


# ============================================================
# STEP 30: CREATE A FUNCTION FOR NEW STUDENT PREDICTION
# ============================================================

def predict_student(
    study_hours,
    attendance,
    motivation,
    previous_gpa,
    internet_access,
    sleep_hours
):
    """
    Predict both CGPA and performance class for a new student.

    internet_access:
    1 = Yes
    0 = No
    """

    new_student = pd.DataFrame([{
        "study_hours": study_hours,
        "attendance": attendance,
        "motivation": motivation,
        "previous_gpa": previous_gpa,
        "internet_access": internet_access,
        "sleep_hours": sleep_hours
    }])

    predicted_cgpa = regression_model.predict(
        new_student
    )[0]

    predicted_class = classification_model.predict(
        new_student
    )[0]

    print("Input Details")
    print("-------------")
    print("Study Hours     :", study_hours)
    print("Attendance      :", attendance)
    print("Motivation      :", motivation)
    print("Previous GPA    :", previous_gpa)
    print("Internet Access :", internet_access)
    print("Sleep Hours     :", sleep_hours)

    print("\nPredicted Outputs")
    print("-----------------")
    print("Predicted CGPA       :", round(predicted_cgpa, 2))
    print("Predicted Performance:", predicted_class)
````
