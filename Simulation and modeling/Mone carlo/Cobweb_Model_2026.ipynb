# **Demand**

---

## **1. What is Demand?**

Demand refers to the **willingness** and **ability** of consumers to buy a good or service at a specific **price** and **time**.

### **Simple Formula**
**Demand = Want + Money + Time**

If any one of these is missing â†’ **No demand exists**.

---

## **2. Key Features of Demand**

- Consumers must **want** the product.
- Consumers must have **money** (purchasing power).
- Demand is measured over a **time period**.
- Price strongly affects demand.
- Change in price â†’ change in quantity demanded.

---

## **3. Examples of Demand**

### **ðŸ“± Example 1: Smartphone**
- Price drops from **Rs. 30,000 â†’ Rs. 25,000**
- More buyers
- **Demand increases**

### **âš¡ Example 2: Electricity**
- Winter â†’ increased heater use  
- **Demand increases**

### **ðŸŒ Example 3: WordPress Hosting**
- Students learning WordPress increases  
- **Demand for hosting increases**

### **ðŸ¥› Example 4: Milk**
- Price rises from **Rs. 80 â†’ Rs. 120**
- People reduce consumption  
- **Demand decreases**

### **ðŸš• Example 5: Taxi/Cab**
- Rainy days â†’ more people need taxis  
- **Demand increases**

---

## **4. Law of Demand**

> **When price increases â†’ demand decreases**  
> **When price decreases â†’ demand increases**

This relationship is **inverse**.

---

## **5. Factors Affecting Demand**

- **Price of the product**
- **Income level**
- **Consumer taste & preferences**
- **Population size**
- **Weather/Season changes**

---

## **6. Summary**

Demand explains how much quantity consumers are **willing + able** to buy at a given price.  
It depends on price, income, taste, population, and time conditions like weather or seasons.

---



```python
import matplotlib.pyplot as plt

# Improved Demand Data
# Higher price â†’ lower quantity (smooth fall)
price = [40000, 36000, 32000, 28000, 24000, 20000, 16000]   # Price decreasing
quantity = [30, 50, 80, 110, 150, 190, 230]                 # Quantity increasing

# Plotting the Demand Curve
plt.figure(figsize=(8,5))

# Plot demand curve
plt.plot(price, quantity, marker='o', linewidth=2)

# Make graph fall from TOP-LEFT to BOTTOM-RIGHT
plt.gca().invert_xaxis()   # price high on left â†’ low on right
plt.gca().invert_yaxis()   # quantity high on left â†’ low on right

plt.title("Improved Demand Curve (Top-Left to Bottom-Right)")
plt.xlabel("Price (Rs.)")
plt.ylabel("Quantity Demanded")
plt.grid(True)

plt.show()
```

# **Supply**

---

## **1. What is Supply?**

Supply refers to the **willingness** and **ability** of producers or sellers to offer a certain quantity of a product for sale at different **prices** and within a specific **time period**.

In simple words:

**Supply = Producerâ€™s Willingness + Ability + Time**

If producers are not willing or not able to produce at a certain price, supply does not exist.

---

## **2. Law of Supply**

The **Law of Supply** states:

> **When the price of a product increases, its supply increases.**  
> **When the price decreases, its supply decreases.**

This means the relationship between **price** and **quantity supplied** is **direct (positive)**.

---

## **3. Key Characteristics of Supply**

- Producers supply **more** when prices rise (profit motive).  
- Producers supply **less** when prices fall.  
- Supply is influenced by:
  - Cost of production  
  - Technology  
  - Number of producers  
  - Tax and subsidies  
  - Future price expectations  

---

## **4. Example: Supply of Smartphones**

Below is a simple example dataset showing how supply changes with price.

| Price (Rs.) | Quantity Supplied |
|-------------|-------------------|
| 20000       | 40                |
| 22000       | 55                |
| 24000       | 70                |
| 26000       | 90                |
| 28000       | 120               |
| 30000       | 160               |

### **Interpretation**
- At **Rs. 20,000**, only **40 units** are supplied.
- At **Rs. 30,000**, supply rises to **160 units**.
- Higher price motivates producers to increase production.
- This matches the **Law of Supply**.

---

## **5. Summary**

Supply describes how much quantity producers are **willing and able** to bring to the market at different price levels.  
As price increases, supply increases; as price decreases, supply decreases.

---


```python
import matplotlib.pyplot as plt

# --- Demand Data (Price â†“, Demand â†‘) ---
price = [30000, 28000, 26000, 24000, 22000, 20000]   # Prices in Rs.
quantity = [50, 65, 80, 100, 130, 170]               # Quantity demanded

# --- Create Figure ---
plt.figure(figsize=(9,6))

# --- Plot the Demand Curve ---
plt.plot(price, quantity, marker='o', markersize=8, linewidth=2, linestyle='-')

# Invert x-axis so the curve slopes DOWNWARD (topâ†’bottom)
plt.gca().invert_xaxis()

# --- Add Point Labels ---
for p, q in zip(price, quantity):
    plt.text(p, q+3, f"{q}", ha='center', fontsize=9)

# --- Titles and Labels ---
plt.title("Demand Curve: Price vs Quantity Demanded", fontsize=14)
plt.xlabel("Price (Rs.)", fontsize=12)
plt.ylabel("Quantity Demanded", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

# **Market Equilibrium: Demand and Supply**

## **1. What is Market Equilibrium?**

Market equilibrium occurs when **quantity demanded (QD)** equals **quantity supplied (QS)** at a particular **price**.

At this point:

- The market is balanced  
- No shortage  
- No surplus  
- Buyers and sellers agree on price and quantity  

This price is called the **Equilibrium Price (Pe)**  
This quantity is called the **Equilibrium Quantity (Qe)**  

---

## **2. How Equilibrium Works**

- If **Price > Equilibrium Price** â†’ Supply > Demand â†’ *Surplus*
- If **Price < Equilibrium Price** â†’ Demand > Supply â†’ *Shortage*
- Market naturally adjusts toward the equilibrium point.

---

## **3. Example Data (Demand & Supply)**

| Price (Rs.) | Quantity Demanded (QD) | Quantity Supplied (QS) |
|-------------|-------------------------|--------------------------|
| 20000       | 170                     | 40                       |
| 22000       | 130                     | 55                       |
| 24000       | 100                     | 70                       |
| 26000       | 80                      | 90                       |
| 28000       | 65                      | 120                      |
| 30000       | 50                      | 160                      |

### **Finding Equilibrium**
- Demand = Supply at **Price = Rs. 26000**  
- At Rs. 26000:  
  - QD = 80  
  - QS = 90  
  - *Closest intersection (approximate equilibrium)*  

Since exact intersection is between **Rs. 24000â€“26000**,  
we take **Rs. 25000 (approx.)** as equilibrium in real markets.

---

## **4. Market Result**
- **Equilibrium Price (Pe): ~ Rs. 25000**  
- **Equilibrium Quantity (Qe): ~ 85 units**

This is the point where demand and supply curves intersect.

---


```python
import matplotlib.pyplot as plt

# Demand and Supply Data
price = [20000, 22000, 24000, 26000, 28000, 30000]

quantity_demand = [170, 130, 100, 80, 65, 50]      # Demand Curve
quantity_supply = [40, 55, 70, 90, 120, 160]       # Supply Curve

# Approximate equilibrium values
eq_price = 25000
eq_quantity = 85

# Plotting
plt.figure(figsize=(9,6))

plt.plot(price, quantity_demand, marker='o', label="Demand", linestyle='-')
plt.plot(price, quantity_supply, marker='o', label="Supply", linestyle='-')

# Plot equilibrium point
plt.scatter(eq_price, eq_quantity, color='red', s=100, label="Equilibrium Point")

plt.title("Market Equilibrium: Demand vs Supply")
plt.xlabel("Price (Rs.)")
plt.ylabel("Quantity")

plt.grid(True)
plt.legend()
plt.show()
```

# **Shortage and Surplus in a Market**

## 1. What is Shortage?
A **shortage** occurs when:
### **Quantity Demanded (QD) > Quantity Supplied (QS)**  
This normally happens when:
- Price is **too low**
- People want to buy more than producers can supply

---

## 2. What is Surplus?
A **surplus** occurs when:
### **Quantity Supplied (QS) > Quantity Demanded (QD)**  
This normally happens when:
- Price is **too high**
- Producers supply more than consumers want

---

## 3. Examples Used in Python Graphs
### **Example A: Milk Market (Shortage)**
- At low price (Rs. 40), demand for milk is high.
- Farmers cannot supply enough milk.
âž¡ï¸ **Shortage**

### **Example B: Rice Market (Surplus due to Good Weather)**
- Good weather = high crop production.
- Rice supply increases.
- But demand remains stable.
âž¡ï¸ **Surplus**

---


```python
#MILK MARKET (SHORTAGE GRAPH)

'''
At low price, consumers want more milk.
Farmers cannot match demand â†’ SHORTAGE
Red region highlights shortage.
'''



import matplotlib.pyplot as plt

# Milk market example (Shortage)
price = [40, 50, 60, 70, 80]

# Quantity demanded decreases as price increases
milk_demand = [100, 90, 70, 50, 40]

# Quantity supplied increases with price
milk_supply = [40, 50, 60, 70, 90]

# At low price (40â€“50), demand > supply â†’ Shortage
shortage_prices = [40, 50]
shortage_qd = [100, 90]
shortage_qs = [40, 50]

plt.figure(figsize=(9,6))
plt.plot(price, milk_demand, marker='o', label="Milk Demand")
plt.plot(price, milk_supply, marker='o', label="Milk Supply")
plt.fill_between(shortage_prices, shortage_qs, shortage_qd, color='red', alpha=0.3, label="Shortage Area")

plt.title("Milk Market: Shortage (QD > QS)")
plt.xlabel("Price (Rs.)")
plt.ylabel("Quantity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#RICE MARKET (SURPLUS GRAPH)

'''
Good weather = large harvest
Supply becomes higher than demand
Green region shows SURPLUS
'''


# Rice market example (Surplus due to good weather)
price = [20, 25, 30, 35, 40]

# Normal demand pattern
rice_demand = [90, 80, 70, 60, 50]

# Good weather increases supply at all price levels
rice_supply = [100, 120, 140, 150, 160]

# Surplus occurs where QS > QD (all points here)
plt.figure(figsize=(9,6))
plt.plot(price, rice_demand, marker='o', label="Rice Demand")
plt.plot(price, rice_supply, marker='o', label="Rice Supply")

# Highlight surplus region
plt.fill_between(price, rice_demand, rice_supply, where=(np.array(rice_supply) > np.array(rice_demand)),
                 color='green', alpha=0.3, label="Surplus Area")

plt.title("Rice Market: Surplus Due to Good Weather (QS > QD)")
plt.xlabel("Price (Rs.)")
plt.ylabel("Quantity")
plt.legend()
plt.grid(True)
plt.show()
```

# **Distributed Lag Model (DLM)**

---

## **1. What is a Distributed Lag Model?**

A **Distributed Lag Model (DLM)** is a time-series model where the **current value** of a dependent variable (Y) depends not only on the **current value** of an independent variable (X), but also on the **past (lagged) values** of X.

In simple words:

> **Today's outcome is influenced by todayâ€™s X and past X values from previous years.**

This model captures **delayed effects** of one variable on another.

---

## **2. Mathematical Form of Distributed Lag Model**

A general *k-lag* distributed lag model is:

\[
Y_t = a + b_0X_t + b_1X_{t-1} + b_2X_{t-2} + \cdots + b_kX_{t-k} + e_t
\]

Where:
- \( Y_t \) = dependent variable at time t  
- \( X_t, X_{t-1}, X_{t-2} \) = current and past values of X  
- \( b_0, b_1, \dots b_k \) = lag coefficients  
- \( e_t \) = error term  

### **Interpretation**
- \( b_0 \): immediate (current-year) effect  
- \( b_1 \): 1-year delayed effect  
- \( b_2 \): 2-year delayed effect  
- ...
- Lag effects can weaken or remain strong depending on the phenomenon.

---

## **3. Why Use Distributed Lag Models?**

Some variables do **not produce immediate effects**.

DLMs help measure:
- partial adjustment  
- delayed impact  
- cumulative long-run effects  

Very useful in:
- Agriculture  
- Economics  
- Environmental science  
- Finance  
- Marketing

---

## **4. Example 1: Rice Production Depends on Past Rainfall**

Let:
- \( P_t \) = rice production in year t  
- \( R_t \) = rainfall this year  
- Soil moisture, seed quality, groundwater are influenced by past rainfall  

Distributed lag model:

\[
P_t = a + b_0R_t + b_1R_{t-1} + b_2R_{t-2}
\]

### **Interpretation**
- \( b_0R_t \): rainfall this year affects crop irrigation  
- \( b_1R_{t-1} \): last yearâ€™s rainfall affects soil fertility  
- \( b_2R_{t-2} \): older rainfall affects long-term groundwater  

Thus, **production this year depends on rainfall over multiple years**, not only the current year.

---

## **5. Example 2: Demand for Smartphones Responds Slowly to Advertising**

Let:
- \( D_t \): demand in year t  
- \( A_t \): advertising expenditure  

Consumers may take time to react to ads.

\[
D_t = a + b_0A_t + b_1A_{t-1} + b_2A_{t-2}
\]

### **Meaning**
- Ads this year create immediate demand  
- Ads last year still influence brand image  
- Ads from 2 years ago may continue to impact loyalty  

This is a **marketing distributed lag**.

---

## **6. Example 3: Fertilizer Use Affects Crop Yield Over Several Seasons**

Let:
- \( Y_t \) = yield in year t  
- \( F_t \) = fertilizer applied in year t  

\[
Y_t = a + b_0F_t + b_1F_{t-1} + b_2F_{t-2} + b_3F_{t-3}
\]

### **Interpretation**
- Fertilizer improves soil nutrients gradually  
- Effects spread over multiple cropping seasons  
- DLM captures long-run soil improvement  

---

## **7. Relation Between Distributed Lag & Demandâ€“Supply**

Yes, DLM fits both:

### **Demand Side**
- Past income influences current consumption  
- Past prices influence purchasing habits  

### **Supply Side**
- Agricultural supply depends on past:
  - rainfall  
  - investment  
  - technology  
  - crop cycle  

Thus, **distributed lag models are natural in demand and supply forecasting**.

---

## **8. Relation Between Distributed Lag & Cobweb Model**

The **Cobweb Model** assumes:
- Supply today depends on past prices  
- Because production decisions have time lag

If farmers also consider **multiple past prices**, then:

\[
QS_t = c + d_0P_{t-1} + d_1P_{t-2}
\]

This becomes a **Distributed Lag Supply Model**, extending the cobweb theory.

### **Key Point**
Cobweb = lagged supply  
Distributed lag = *multi-year* lagged supply  

So, **DLM broadens and strengthens the cobweb model** by considering multiple past periods.

---

## **9. Summary**

- A **Distributed Lag Model** explains how current values depend on current + multiple past values of another variable.  
- Useful for systems with **delayed or gradual influence**.  
- Fits well with **demand**, **supply**, and especially **agricultural markets**.  
- Integrates naturally with **Cobweb Theory**, which is based on lagged supply responses.

# **MORE ON ----------------------------       Lag Models and Distributed Lag Models**

---

## **1. What is a Lag Model?**

A **Lag Model** (Time Lag Model) is a time-series model where the **current value** of a variable depends on its **past (lagged) values**.

### **General Form**
\[
Y_t = a + bY_{t-1} + e_t
\]

Where:
- \( Y_t \) = value this year  
- \( Y_{t-1} \) = value last year  
- \( b \) = lag effect coefficient  
- \( e_t \) = error term  

### **Simple Meaning**
Todayâ€™s outcome depends partly on yesterdayâ€™s outcome.

---

## **2. Example of a Lag Model (Rice Production)**

Let:

- \( P_{2023} \) = 800 tons  
- \( P_{2024} \) = depends on rainfall + technology + **last yearâ€™s production**  
- \( P_{2025} \) = depends on previous year again  

### **Lag Relationship**
\[
P_{2024} = a + bP_{2023}
\]
\[
P_{2025} = a + bP_{2024}
\]

### **Interpretation**
- Good production in 2023 â†’ better resources and seeds â†’ increases 2024 output  
- Poor production in 2024 â†’ lower investment â†’ reduces 2025 output  

This is a **one-period lag model**.

---

## **3. What is a Distributed Lag Model?**

A **Distributed Lag Model** uses **multiple lags**, meaning the current value depends on several past periods.

### **General Form**
\[
Y_t = a + b_0X_t + b_1X_{t-1} + b_2X_{t-2} + \cdots + b_kX_{t-k} + e_t
\]

### **Meaning**
Current output depends on:
- this yearâ€™s factor  
- last yearâ€™s factor  
- 2 years earlier  
- etc.

---

## **4. Example of Distributed Lag (Rice Production)**

Let rice output depend on rainfall:

\[
P_t = a + b_0R_t + b_1R_{t-1} + b_2R_{t-2}
\]

Where:
- \( R_t \) = rainfall this year  
- \( R_{t-1} \) = last year  
- \( R_{t-2} \) = rainfall two years ago  

### **Explanation**
- Rainfall this year affects irrigation.  
- Rainfall last year affects soil moisture and seed quality.  
- Rainfall two years ago affects long-run soil fertility.  

This is a **distributed lag model (multiple time lags)**.

---

## **5. Does Lag Theory Fit Demand and Supply?**

### âœ” Yes, many real markets show **lagged behavior**.

### **Lag in Demand**
Demand may depend on:
- previous yearâ€™s income  
- previous consumption habits  
- trend persistence  

Example:
\[
QD_t = a + b QD_{t-1}
\]

---

### **Lag in Supply**
Supply typically has **strong lags** because production takes time.

Examples:
- Agricultural supply today depends on last yearâ€™s climate.  
- Manufacturing supply depends on past investments.  

\[
QS_t = a + b QS_{t-1}
\]

Thus, **lag models fit both demand and supply**, especially agricultural markets.

---

## **6. Connection to the Cobweb Model**

The **Cobweb Model** explains price fluctuations when supply reacts with a **time lag**.

### **Why Does the Cobweb Model Need Lag?**
Because:
- farmers choose crop quantity **this year** based on **last yearâ€™s price**  
- production cannot be adjusted instantly  

Thus, **lag in supply is the core reason cobweb patterns occur**.

---

## **7. How Lag Models Fit the Cobweb Model**

In the Cobweb Model:

\[
QS_t = c + dP_{t-1}
\]
\[
QD_t = a - bP_t
\]

Producers look at *past price (lag)* to decide supply.  
This creates oscillations:

- If supply reacts too strongly â†’ unstable cobweb  
- If supply reacts moderately â†’ stable cobweb  
- If perfect â†’ constant cycle

### **Cobweb depends on Lagged Supply**
- Decision today = function of yesterdayâ€™s price  
- This is exactly a **lag model** applied to supply  
- And behaves like a **distributed lag** if farmers consider multiple past years  

Therefore:

âœ” Lag Model â†’ Explains delayed response  
âœ” Distributed Lag â†’ Realistic multi-year effects  
âœ” Cobweb Model â†’ Market instability due to lag  

---

## **8. Summary**

- **Lag Model:** current value depends on last yearâ€™s value.  
- **Distributed Lag Model:** current value depends on several past years.  
- **Rice production** is a perfect real-world example.  
- **Demand and supply** both show lag effects.  
- **Cobweb Model** is built entirely on **lagged supply response**, explaining price oscillations in agriculture.

# **Exampe and CODE 1**

```python
import numpy as np
import matplotlib.pyplot as plt

# Years from 2015 to 2025
years = np.arange(2015, 2026)
n_years = len(years)

# Base rainfall pattern to reflect your story:
# 2015: good rainfall
# 2016: bad rainfall
# 2017: recovery, etc.
base_rainfall = np.array([
    1.2,  # 2015 - good
    0.6,  # 2016 - bad
    0.9,  # 2017 - medium
    1.0,  # 2018
    1.1,  # 2019
    0.8,  # 2020
    1.0,  # 2021
    0.9,  # 2022
    1.2,  # 2023 - good again
    0.7,  # 2024 - weaker
    1.0   # 2025 - normal
])

np.random.seed(42)  # for reproducible runs

for run in range(10):
    # Add small random variation to rainfall each run
    rainfall = base_rainfall + np.random.normal(0, 0.1, size=n_years)
    rainfall = np.clip(rainfall, 0.3, 1.5)  # keep within reasonable range

    # Arrays for production and price
    production = np.zeros(n_years)
    price = np.zeros(n_years)

    # Initial values
    production[0] = 100 + 50 * rainfall[0]   # base production
    price[0] = 30 + np.random.uniform(-3, 3) # initial price (e.g., Rs/kg)

    # Model parameters
    # Production_t depends on rainfall_t and previous production
    alpha_prod = 20      # base level
    beta_rain = 60       # effect of rainfall on production
    gamma_prod = 0.4     # persistence from last year's production

    # Price_{t+1} depends on current production_t and current price_t
    # (Here: more production â†’ lower next year's price)
    alpha_price = 10
    beta_prod = -0.1     # negative: higher production â†’ lower price next year
    gamma_price = 0.6    # price persistence

    for t in range(1, n_years):
        # Production this year
        noise_prod = np.random.normal(0, 5)
        production[t] = (
            alpha_prod
            + beta_rain * rainfall[t]
            + gamma_prod * production[t-1]
            + noise_prod
        )

        # Price this year depends on last year's production (lag) and last year's price
        noise_price = np.random.normal(0, 1.5)
        price[t] = (
            alpha_price
            + beta_prod * production[t-1]   # last year's production
            + gamma_price * price[t-1]      # last year's price
            + noise_price
        )

    # Plot each run as a separate graph
    plt.figure(figsize=(7, 4))

    # Scale rainfall so it shows nicely with other series
    rainfall_scaled = rainfall * 100

    plt.plot(years, rainfall_scaled, marker='o', linestyle='--', label="Rainfall Index (scaled)")
    plt.plot(years, production, marker='s', linestyle='-', label="Rice Production")
    plt.plot(years, price, marker='^', linestyle='-', label="Rice Price")

    plt.title(f"Riceâ€“Rainfall Lag Simulation (Run {run + 1})")
    plt.xlabel("Year")
    plt.ylabel("Level (indexed)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

plt.show()
```

# **Rice Production Lag Model: Rainfall â†’ Production â†’ Price (2015â€“2025)**

---

## **1. Concept Overview**

Agricultural markets often show **lag relationships**, meaning:

- **Rainfall in the current year** affects  
  **â†’ Rice production in the current year**

- **Rice production today** affects  
  **â†’ Rice price in the next year**

This happens because:
- Good rainfall improves crop growth.
- Higher production increases supply â†’ lowers next year's price.
- Poor rainfall reduces crop output â†’ increases next year's price.
- Prices also show persistence (last year's price affects this year's price).

Thus, the mechanism is:

\[
\text{Rainfall}_t \rightarrow \text{Production}_t \rightarrow \text{Price}_{t+1}
\]

---

## **2. Why Does This Lag Occur?**

### **Lag in Production**
Agricultural output reacts **immediately** to rainfall because:
- Better rainfall â†’ irrigation, soil moisture, healthy crops  
- Poor rainfall â†’ drought stress, weak crops

### **Lag in Price**
Market prices react **with delay** because:
- Harvest happens once a year  
- Market adjusts only after production is known  
- Next seasonâ€™s price is based on current supply

So, this is a **distributed lag relationship**:
- One lag for production  
- One lag for price  

---

## **3. Realistic Example Story (2015â€“2025)**

### **2015**
- Rainfall: Good  
- Production: High  
- 2016 Price: Falls (because supply was high)

### **2016**
- Rainfall: Bad  
- Production: Low  
- 2017 Price: Rises (because supply was low)

### **2017**
- Rainfall: Medium  
- Production: Moderate  
- 2018 Price: Adjusts accordingly

And this pattern continues until **2025**.

\[

\]

Where:
- Higher last year's production â†’ lowers current price  
  (`eta` is negative)
- Last year's price influences current price level  
- Random shocks simulate market noise

## **1. Concept Overview**

In real markets, the impact of advertising on demand is **not immediate**.

- Some buy after months  
- Some buy next year  
- Some wait for discounts, salary cycle, festival seasons, or new models  

This means **advertising has a delayed (lagged) effect on demand**.

This behavior is modeled using a **Distributed Lag Model (DLM)**.

---

## **2. Distributed Lag Model for Smartphone Demand**

Let:

- \( D_t \) = Demand for smartphones this year  
- \( A_t \) = Advertising expenditure this year  
- \( A_{t-1} \), \( A_{t-2} \) = advertising expenditures in previous years  

The model is:

\[
D_t = a + b_0 A_t + b_1 A_{t-1} + b_2 A_{t-2} + e_t
\]

### **Interpretation**
- **\(b_0 A_t\)** â†’ immediate impact of advertising  
- **\(b_1 A_{t-1}\)** â†’ delayed impact from last year's ads  
- **\(b_2 A_{t-2}\)** â†’ long-lasting brand influence from ads two years ago  

Smartphone products (Android, iPhone, etc.) show this behavior strongly.

---

## **3. Why Does Demand Respond Slowly?**

### **Smartphone buying is not impulsive**
- High cost â†’ requires savings  
- People wait for:
  - Dashain, Tihar, Black Friday, New Year offers  
  - EMI schemes  
  - New model launch  
  - Salary bonus  

### **Brand-building takes time**
- Repeated ads shape perception gradually  
- Trust builds slowly  
- Friends/influencer reviews matter  

### **Information processing delay**
Consumers compare:
- Camera quality  
- Battery life  
- Processor speed  
- Warranty  
- Discount vs. value  

This causes a **time lag** between advertisement and actual purchase.

## **4. Practical Example Dataset**

| Year | Advertising Budget (A) | Demand (D) |
|------|--------------------------|------------|
| 2020 | 10 million               | 50,000     |

| 2022 | 15 million               | 60,000     |

### **Observation**
- Demand in 2021 increased partly because of **2020 ads**  
- Demand in 2022 increased because of **2020 + 2021 ads**  
- Demand in 2023 increased because of **2021 + 2022 + 2023 ads**  

This perfectly fits a **distributed lag pattern**.

## **5. How It Looks in Reality**

## **6. Summary**

- Smartphone demand reacts **slowly**, not instantly, to advertising.

- This method is realistic for analyzing:
  - marketing effectiveness  
  - consumer response time  
  - brand-building strategies

```python
import numpy as np
import matplotlib.pyplot as plt

# Years
years = np.arange(2015, 2026)  # 2015â€“2025
n = len(years)

# 1) Advertising pattern (in million Rs.)
# Simulate realistic ups and downs in ad spending
ad_spend = np.array([2, 3, 4, 5, 6, 5, 4, 6, 7, 6, 5], dtype=float)

# 2) Distributed Lag Model parameters for demand
a = 50      # base demand level (thousand units)
b0 = 8      # immediate effect of current advertising
b1 = 5      # 1-year lag effect
b2 = 2      # 2-year lag effect

# 3) Initialize demand array
demand = np.zeros(n)

# We need at least 2 previous years, so start from index 2
# For t = 0 and t = 1, we can use reduced form
demand[0] = a + b0 * ad_spend[0]              # first year (no lags)
demand[1] = a + b0 * ad_spend[1] + b1 * ad_spend[0]  # second year (one lag)

# Add random noise to make it more realistic
np.random.seed(42)

for t in range(2, n):
    noise = np.random.normal(0, 3)  # random shock
    demand[t] = (
        a
        + b0 * ad_spend[t]
        + b1 * ad_spend[t - 1]
        + b2 * ad_spend[t - 2]
        + noise
    )

# 4) Plotting Advertising and Demand over time
plt.figure(figsize=(9, 5))

# Scale advertising to make it visible with demand
ad_scaled = ad_spend * 10  # just for visualization

plt.plot(years, ad_scaled, marker='o', linestyle='--', label="Advertising (scaled)")
plt.plot(years, demand, marker='s', linestyle='-', label="Smartphone Demand")

plt.title("Demand for Smartphones Responds Slowly to Advertising\n(Distributed Lag Model)")
plt.xlabel("Year")
plt.ylabel("Level (index units)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Distributed Lag Model (DLM): Definition, Need, Types, and Examples**

---

## **1. What is a Distributed Lag Model?**

A **Distributed Lag Model (DLM)** is an econometric/time-series model in which the **current value** of a dependent variable (Yâ‚œ) depends on both:

1. The **current value** of an independent variable (Xâ‚œ), and  
2. The **lagged (past) values** of that independent variable (Xâ‚œâ‚‹â‚, Xâ‚œâ‚‹â‚‚, ... Xâ‚œâ‚‹â‚–)

### **General Form**
\[
Y_t = a + b_0X_t + b_1X_{t-1} + b_2X_{t-2} + \cdots + b_kX_{t-k} + e_t
\]

Where:
- \(Y_t\) = dependent variable  
- \(X_t, X_{t-1}, \ldots\) = current and past values of X  
- \(b_0, b_1, \ldots, b_k\) = distributed lag coefficients  
- \(e_t\) = random error  

---

## **2. Why Distributed Lag Models? (Importance)**

Certain economic effects are **not immediate**, they occur **gradually over time**.

### **Reasons:**
- Consumers respond slowly (habits, adjustment lag)
- Producers adjust slowly (inventory, crop cycle, investment)
- Weather or environment has multi-year impact
- Policies take time to show full effect
- Advertising creates delayed brand influence
- Income changes affect future consumption

### **In simple words:**
> **DLM captures the delayed or accumulated influence of past events on current outcomes.**

---

## **3. Applications of Distributed Lag Models**

### âœ” **Agriculture**
- Rainfall over last 2â€“3 years affects todayâ€™s crop yield  
- Fertilizer use in multiple years affects soil quality  

### âœ” **Marketing**
- Advertising effect spreads across future periods  
- Brand recall from old ads still affects demand  

### âœ” **Macroeconomics**
- GDP responds slowly to government expenditure  
- Inflation responds slowly to money supply changes  

### âœ” **Finance**
- Stock returns influenced by past market volatility  
- Interest rates affecting investments over time  

### âœ” **Environmental Economics**
- Pollution affects health outcomes gradually  

---

## **4. Types of Distributed Lag Models Based on Stability**

Depending on how the coefficients \(b_0, b_1, b_2, \ldots\) behave over time, a DLM can be:

---

## **A. Equilibrium Distributed Lag Model**

Occurs when lag effects eventually settle to a stable long-run relationship.

### **Example**
Rice demand depends on:
- Current year income  
- Last year's income  
- Two years ago income  

\[
D_t = a + 0.3Y_t + 0.2Y_{t-1} + 0.1Y_{t-2}
\]

The coefficients decrease gradually â†’ stable long-term equilibrium.

### **Interpretation**
- Current income has strongest effect  
- Past income has weaker effect  
- Long-run impact stabilizes  

---

## **B. Convergent Distributed Lag Model**

Convergent occurs when coefficients **decrease over time**, meaning past values have **less influence**.

\[
Y_t = a + 0.5X_t + 0.3X_{t-1} + 0.1X_{t-2}
\]

### **Characteristics**
- Lag impact becomes smaller  
- Stable  
- System moves to equilibrium smoothly  

### **Example**
Demand for smartphones:
- Advertising today has high impact  
- Advertising last year has small impact  
- Advertising two years ago has very small effect  

---

## **C. Divergent Distributed Lag Model**

Coefficients **increase** or **do not decrease**, causing instability.

\[
Y_t = a + 0.3X_t + 0.4X_{t-1} + 0.5X_{t-2}
\]

### **Characteristics**
- Past values have more influence than current  
- System becomes unstable  
- Shocks become bigger over time  

### **Example**
Agricultural supply:
- Past good harvest encourages large investment  
- Future supply expands aggressively  
- If oversupply occurs, prices crash â†’ instability  

(Links to **divergent cobweb model**)

---

## **5. Distributed Lag and Cobweb Model**

The **Cobweb Model** depends on lagged supply:

\[
Q^S_t = c + dP_{t-1}
\]

If farmers also consider **multiple past prices**, then cobweb becomes a **distributed lag supply model**:

\[
Q^S_t = c + d_0P_{t-1} + d_1P_{t-2} + d_2P_{t-3}
\]

### **Meaning:**
- Supply depends on several past prices  
- Can result in:
  - **Convergent cycles**  
  - **Divergent cycles**  
  - **Equilibrium**  

---

## **6. Summary**

- DLM measures **delayed and accumulated impacts**.
- Useful where effects occur slowly (agriculture, advertising, income).
- Three types of DLM stability:
  - **Equilibrium** (balanced)
  - **Convergent** (stable)
  - **Divergent** (unstable)
- Cobweb model is a real-world example of DLM in supply response.

```python
import numpy as np
import matplotlib.pyplot as plt

# ----- Parameters from the question -----
a_d = 12.4   # demand intercept
b_d = 1.2    # demand slope (Qd = a_d - b_d * P)

c_s = 1.0    # supply intercept
d_s = 0.9    # supply slope (Qs = c_s + d_s * P_prev)

# Inverse demand and inverse supply (for plotting straight lines)
def P_d(Q):
    return (a_d - Q) / b_d          # from Qd = a_d - b_d P

def P_s(Q):
    return (Q - c_s) / d_s          # from Qs = c_s + d_s P

# ----- Equilibrium (static) -----
# Solve 12.4 - 1.2 P = 1.0 + 0.9 P
P_eq = 11.4 / 2.1
Q_eq = 12.4 - 1.2 * P_eq

print("Equilibrium price:", P_eq)
print("Equilibrium quantity:", Q_eq)

# ----- Set up quantity range for curves -----
Q_min = 0
Q_max = 12
Q_vals = np.linspace(Q_min, Q_max, 200)

# ----- Cobweb simulation -----
T = 12                     # number of periods
P = np.zeros(T + 1)
Q_s_path = np.zeros(T + 1)

# Choose an initial price away from equilibrium
P[0] = 2.0                 # you can change this
Q_s_path[0] = c_s + d_s * P[0]

for t in range(T):
    # Given P[t], supply this period:
    Q_s_t = c_s + d_s * P[t]      # point on supply curve
    Q_s_path[t] = Q_s_t

    # New price from demand at that quantity: Qs_t = 12.4 - 1.2 P[t+1]
    P_next = (a_d - Q_s_t) / b_d
    P[t + 1] = P_next

# Last period's supply quantity for plotting
Q_s_path[T] = c_s + d_s * P[T]

# ----- Plot -----
plt.figure(figsize=(8, 6))

# Demand and supply curves
plt.plot(Q_vals, P_d(Q_vals), label="Demand (Q = 12.4 - 1.2P)")
plt.plot(Q_vals, P_s(Q_vals), label="Supply (Q = 1.0 + 0.9P)")

# Cobweb path: alternating horizontal and vertical moves
for t in range(T):
    # From supply point (Q_s_t, P[t]) horizontally to demand (same Q, new P[t+1])
    Q_s_t = Q_s_path[t]
    P_t = P[t]
    P_next = P[t + 1]

    # horizontal: from (Q_s_t, P_t) to (Q_s_t, P_next)
    plt.plot([Q_s_t, Q_s_t], [P_t, P_next])

    # vertical: from (Q_s_t, P_next) to (Q_s_{t+1}, P_next)
    Q_s_next = Q_s_path[t + 1]
    plt.plot([Q_s_t, Q_s_next], [P_next, P_next])

# Mark equilibrium
plt.scatter([Q_eq], [P_eq])
plt.text(Q_eq, P_eq, "  Equilibrium", va="bottom")

plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title("Cobweb Model for Q = 12.4 - 1.2P  and  Q = 1.0 + 0.9Pâ‚â‚‹â‚â‚Ž")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
