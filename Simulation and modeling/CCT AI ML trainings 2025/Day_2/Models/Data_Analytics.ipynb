**Required Libraries:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Enable visual style
sns.set(style="whitegrid")
```

**Load the Dataset**

```python
# Load the CSV file
df = pd.read_csv('/content/Salary Data.csv', na_values=["NA", "N/A", "", "null", "NaN", "?"])

# Show the first few rows
print("ðŸ”¹ First 5 rows of the dataset:")
print(df.head())
```

**Data Cleaning**

```python
# Check for missing values
print("\nðŸ”¹ Missing values in each column:")
print(df.isnull().sum())

# Drop rows with missing values (optional â€” or you can fill them)
df.dropna(inplace=True)

# Check for duplicates
print("\nðŸ”¹ Number of duplicate rows:", df.duplicated().sum())

# Remove duplicates if any
df.drop_duplicates(inplace=True)

# Standardize categorical columns (optional cleaning)
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df['Education Level'] = df['Education Level'].str.strip().str.title()
df['Job Title'] = df['Job Title'].str.strip().str.title()
```

**Descriptive Statistics**

```python
print("\nðŸ”¹ Descriptive statistics for numerical columns:")
print(df.describe())

print("\nðŸ”¹ Value counts for categorical columns:")
print("\nGender:\n", df['Gender'].value_counts())
print("\nEducation:\n", df['Education Level'].value_counts())
print("\nJob Title:\n", df['Job Title'].value_counts())
```

**Grouped Analysis**

```python
# Average salary by gender
print("\nðŸ”¹ Average salary by Gender:")
print(df.groupby('Gender')['Salary'].mean())

# Average salary by Education
print("\nðŸ”¹ Average salary by Education:")
print(df.groupby('Education Level')['Salary'].mean())

# Average salary by Job title
print("\nðŸ”¹ Average salary by Job Title:")
print(df.groupby('Job Title')['Salary'].mean())
```

**Visualizations**

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Salary'], bins=20, color='teal', edgecolor='black')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

**Gender Distribution**

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

**Education Levels**

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Education', palette='muted')
plt.title("Education Level Distribution")
plt.xlabel("Education")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

**Bar Plot: Salary vs Gender**

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x='Gender', y='Salary', palette='Set2')
plt.title("Salary Distribution by Gender")
plt.show()
```

**Bar Plot: Salary vs Education**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Education', y='Salary', palette='Set3')
plt.title("Salary Distribution by Education")
plt.xticks(rotation=45)
plt.show()
```

**Scatter Plot: Salary vs Years of Experience**

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='Years of Experience', y='Salary', hue='Gender', palette='Set1')
plt.title("Salary vs. Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend(title='Gender')
plt.show()
```

**Scatter Plot: Salary vs Age**

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='Age', y='Salary', hue='Education', palette='Set1')
plt.title("Salary vs. Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(title='Education', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Correlation Heatmap (for numerical columns)**

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Age', 'Years of Experience', 'Salary']].corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
