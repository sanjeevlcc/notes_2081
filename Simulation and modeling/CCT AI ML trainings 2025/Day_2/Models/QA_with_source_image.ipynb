```python
# langchain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

# langchain-chroma
from langchain_chroma import Chroma
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain

import torch
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=0)
chunked_documents = text_splitter.split_documents(data)
```

```python
database = Chroma.from_documents(chunked_documents, HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
```

```python
chunked_documents
```

```python
retriever = database.as_retriever()
```

```python
query = "What is the term of agreement?"
docs = database.similarity_search(query)
docs
```

```python
import transformers
from transformers import AutoTokenizer, pipeline, AutoModelForSeq2SeqLM
import torch
checkpoint="MBZUAI/LaMini-T5-738M"
#We will be using the most stable modelk (Either llama 3 or mixtral)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
base_model = AutoModelForSeq2SeqLM.from_pretrained(
    checkpoint,
    # device_map="auto",
    torch_dtype=torch.float32
)
```

```python
pipe=pipeline(
        'text2text-generation',
        model=base_model,
        tokenizer=tokenizer,
        temperature=0.2,
        max_length=1024,
        do_sample=True,
        top_p=0.95
    )
llm=HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.chains import RetrievalQA
qa=RetrievalQA.from_chain_type(
      llm=llm,
      chain_type="stuff",
      retriever=retriever,
      return_source_documents=True
    )
```

```python
def ragsys(query):
  # retriever
  retriever = database.as_retriever()
  ret_chunk = retriever.invoke(query)
  source = ret_chunk[0].metadata.get('source')
  image = Image.open(source)
  my_image = plt.imshow(image)
  plt.axis('off')
  plt.show()
  # rag
  qa=RetrievalQA.from_chain_type(
      llm=llm,
      chain_type="stuff",
      retriever=retriever,
      return_source_documents=True
    )
  output = qa({"query": question})
  print(output['result'])
```

```python
def highlight_text(image, text, color=(0, 255, 0)):
    # Find the bounding box of the text in the image
    (x, y, w, h), _ = cv2.textLayout(text, cv2.FONT_HERSHEY_SIMPLEX, 1, thickness=2)
    # Draw a rectangle around the text
    cv2.rectangle(image, (x, y - h), (x + w, y), color, 2)

```

```python
# Iterate over text chunks and highlight them in the image
for chunk in ret_chunk:
    if chunk.strip():  # Check if the chunk is not empty
        highlight_text(my_image, chunk)
```
