```python
cleaned_df = df.copy()
```

```python
print("Copied raw data into cleaned_df for cleaning.")
cleaned_df.head()
```

```python
cleaned_df = cleaned_df.drop_duplicates()

print("After removing duplicates:")
print(cleaned_df.shape)
```

```python
cleaned_df
```

```python

cleaned_df["Quantity"] = cleaned_df["Quantity"].fillna(1)

print("Missing Quantity fixed.")
print(cleaned_df["Quantity"].isnull().sum())
```

```python
cleaned_df
```

```python
cleaned_df["City"] = cleaned_df["City"].fillna("Unknown")

print("Missing City fixed.")
print(cleaned_df["City"].isnull().sum())
```

```python
cleaned_df
```

```python
cleaned_df["Unit Price"] = cleaned_df["Unit Price"].fillna(2500)

print("Missing Unit Price fixed.")
print(cleaned_df["Unit Price"].isnull().sum())
```

```python
cleaned_df
```

```python
cleaned_df["Quantity"] = cleaned_df["Quantity"].abs()

print("Negative Quantity corrected.")
cleaned_df[cleaned_df["Order ID"] == 1022]
```

```python
cleaned_df
```

```python
cleaned_df["City"] = cleaned_df["City"].str.strip().str.title()
cleaned_df["Category"] = cleaned_df["Category"].str.strip().str.title()
cleaned_df["Payment Method"] = cleaned_df["Payment Method"].str.strip().str.title()
cleaned_df["Product"] = cleaned_df["Product"].str.strip().str.title()
cleaned_df["Customer Name"] = cleaned_df["Customer Name"].str.strip().str.title()
```

```python
print("Text normalization completed.")
cleaned_df.head()
```

```python
cleaned_df["Date"] = pd.to_datetime(cleaned_df["Date"])

print("Date column converted successfully.")
cleaned_df.info()
```

```python
cleaned_df["Total Sales"] = cleaned_df["Quantity"] * cleaned_df["Unit Price"]

print("Total Sales column added.")
cleaned_df.head()
```

```python
print("Final Cleaned Dataset Shape:")
print(cleaned_df.shape)
```

```python
print("\nMissing Values After Cleaning:")
print(cleaned_df.isnull().sum())
```

```python
print("Basic Descriptive Statistics for Total Sales:")

total_sales = cleaned_df["Total Sales"].sum()
mean_sales = cleaned_df["Total Sales"].mean()
median_sales = cleaned_df["Total Sales"].median()
mode_sales = cleaned_df["Total Sales"].mode()
minimum_sales = cleaned_df["Total Sales"].min()
maximum_sales = cleaned_df["Total Sales"].max()
order_count = cleaned_df["Total Sales"].count()
```

```python
print("Total Orders  :", order_count)
print("Total Sales   :", total_sales)
print("Mean Sales    :", mean_sales)
print("Median Sales  :", median_sales)
print("Mode Sales    :", list(mode_sales))
print("Minimum Sales :", minimum_sales)
print("Maximum Sales :", maximum_sales)

```

```python
cleaned_df[["Quantity", "Unit Price", "Total Sales"]].describe()
```

```python

category_summary = cleaned_df.groupby("Category")["Total Sales"].sum().reset_index()

category_summary = category_summary.sort_values(by="Total Sales", ascending=False)

print("Category-wise Sales Summary:")
category_summary
```

```python
city_summary = cleaned_df.groupby("City")["Total Sales"].sum().reset_index()

city_summary = city_summary.sort_values(by="Total Sales", ascending=False)

print("City-wise Sales Summary:")
city_summary
```

```python

category_pivot = pd.pivot_table(
    cleaned_df,
    values="Total Sales",
    index="Category",
    aggfunc="sum"
)

category_pivot
```

```python
plt.figure(figsize=(8,5))
plt.bar(category_summary["Category"], category_summary["Total Sales"])
plt.title("Category-wise Sales Analysis")

plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(city_summary["City"], city_summary["Total Sales"])
plt.title("City-wise Sales Performance")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
payment_summary = cleaned_df.groupby("Payment Method")["Total Sales"].sum().reset_index()

plt.figure(figsize=(6,6))
plt.pie(
    payment_summary["Total Sales"],
    labels=payment_summary["Payment Method"],
    autopct="%1.1f%%"
)
plt.title("Payment Method Sales Share")
plt.show()
```

```python

daily_sales = cleaned_df.groupby("Date")["Total Sales"].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(daily_sales["Date"], daily_sales["Total Sales"], marker="o")
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```
